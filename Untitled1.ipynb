```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from db import Records, conn_dict
```

```python
q = '''
     select anio, count(*)
     from hechos
     group by anio
     order by anio;
 '''
```

```python
a = Records(conn_dict, table='hechos', query=q)
```

```python
rec = a.pds_df()
```

```python
sns.lmplot('anio', 'count', data=rec, fit_reg=False)
```

```python
sns.jointplot(x="anio", y="count", data=rec, kind="kde");
```
